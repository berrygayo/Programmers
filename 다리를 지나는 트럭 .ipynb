### 변수설명 
+ passing_weight : 각 트럭이 다리에 오르는 순간부터 얼마의 시간이 흘렀는지를 나타내는 리스트 
+ passed: 지나간 트럭 
+ passing: 지나는 중인 트럭 

+ i: 가장 선두에서 지나는 트럭의 index :: bridge_length만큼 다리위에 있었는지 확인하기 위함 
+ j : 가장 최근 다리에 올라간 트럭으 index ::passing_weight에 indexing하기 위함 


```python
def solution(bridge_length, weight, truck_weights) :
    answer=0
    
    truck_weights = truck_weights[::-1] #순서 바꿔줌 pop()으로 맨 앞에 7부터 꺼내기 위해서
    n = len(truck_weights)
    
    
    # 위에 변수 설정 
    passing_weight = [0]*n
    passed = []
    passing = []
 
    i = 0
    j = -1
    
    while len(passed)<n : #다 도착하면 멈춤 
        if len(truck_weights)>0 and sum(passing) + truck_weights[-1] <= weight :
            #trucj_weights에 아직 지나갈 트럭들이 남아있고, 
            # 지나가는 차의 무게랑 다음 들어올 트럭의 무게의 합이 weight 보다 작으면 
            passing.append(truck_weights.pop())
            j+=1 #위에서 passing에 하나 올렸으니까 j +1 
        passing_weight[:j+1] = [passing_weight[z]+1 for z in range(j+1)] #시간 체크 
 
        if passing_weight[i] == bridge_length :
            passed.append(passing[0])
            passing = passing[1:]
            i+=1
    answer=passing_weight[0]+1
 
    return answer


```
