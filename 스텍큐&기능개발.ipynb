# 스택(Stack)
- 나중에 넣은 데이터를 먼저 반환하도록 설계된 메모리 구조로 Last In Out(LIFO)로 구현됨
- Date의 입력을 Push, 출력을 Pop이라고 함

# 스택(Stack) in 파이썬 
- 파이썬은 리스트를 사용하여 스텍 구조를 활용 
- push를 append(), pop을 pop()를 사용 

```python
a=[1,2,3,4,5]
a.append(10)
a.append(20)
```

```python
a.pop() # 맨 뒤에 10이 나온다
a
```

```python
a.pop() # 맨 뒤에 5가 나온다
a
```

```python
for _ in range(len(a)):
    print(a.pop())
```

```python
# 스택 구조를 활용, 입력된 글자를 역으로 출력 
word=input("Input a word:")
word_list=list(word)
for _ in range(len(word_list)):
    print(word_list.pop())
```

# Queue in 파이썬 
- 파이썬은 리스트를 사용하여 큐 구조를 활용
- put를 append(), get을 pop(0)를 사용 

```python
a=[1,2,3,4,5]
a.append(10)
a.append(20)
a
```

```python
a.pop(0) #0번째 index를 뽑아옴 
```

```python
def solution(progresses, speeds):
    temp=[]
    answer = [1] #첫번재꺼 최소 한개는 배포 되니까 
    
    for i in range(len(progresses)):
        count=0                
        sum=progresses[i]
        while 1:
            sum+=speeds[i]
            count+=1
            
            if sum>=100:
                break;
        temp.append(count)
      
    count2=1
    for i in range(1,len(temp)):
        if temp[i] > temp[i-1]: #뒤에 걸리는 날짜가 앞 날짜보다 더 오래 걸리면 
            count2=1
            answer.append(count2)
        else: # 작다면 앞에 배포하는 날짜에 같이 배포되니가 
            count2+=1
            ##스텍-큐 
            answer.pop() # 그 전의 값을 제거해주고 
            answer.append(count2)# 작은 값들을 다시 넣어준다 
            
    
    return answer


solution([93,30,55],[1,30,5])
```
